```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import BayesianRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
dataset = pd.read_csv("tcd ml 2019-20 income prediction training (with labels).csv")
testset = pd.read_csv("tcd ml 2019-20 income prediction test (without labels).csv")
# Check for Null Data
dataset.isnull().sum()

dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
dataset['Gender'] = dataset['Gender'].fillna('unknown')
dataset['Profession'] = dataset['Profession'].fillna('unknown')
dataset['University Degree'] = dataset['University Degree'].fillna('No')
dataset['Hair Color'] = dataset['Hair Color'].fillna('Unknown')
dataset['Year of Record'].fillna(dataset['Year of Record'].mean(), inplace=True)

#dataset=dataset.drop(['Year of Record'],axis=1)
dataset=dataset.drop(['Profession'],axis=1)
dataset=dataset.drop(['Country'],axis=1)

#testset=testset.drop(['Year of Record'],axis=1)
testset=testset.drop(['Profession'],axis=1)
testset=testset.drop(['Country'],axis=1)

dataset.isnull().sum() #CHECKING FOR NULL
testset.isnull().sum()

#FILLING MISSING VALUES
testset['Age'].fillna(testset['Age'].mean(), inplace=True)
testset['Gender'] = testset['Gender'].fillna('unknown')
testset['University Degree'] = testset['University Degree'].fillna('No')
testset['Hair Color'] = testset['Hair Color'].fillna('Unknown')
testset['Year of Record'].fillna(testset['Year of Record'].mean(), inplace=True)


testset.isnull().sum() #CHECKING FOR NULL

#FILLING MISSING VALUES
dataset['Gender'] = dataset['Gender'].replace(['0','#N/A' ], 'unknown')                   #replacing 0s with categorical values
dataset['University Degree'] = dataset['University Degree'].replace(['0','#N/A'], 'No')
dataset['Hair Color'] = dataset['Hair Color'].replace(['0','#N/A'], 'Unknown')
dataset['Year of Record']=dataset['Year of Record'].replace(['#N/A'],'0')

#FILLING MISSING VALUES
testset['Gender']=testset['Gender'].replace(['0','#N/A'],'unknown')
testset['Gender'].unique #Checking the column values
testset['Body Height [cm]'].unique #Checking the column values
dataset['Year of Record']=dataset['Year of Record'].replace(['#N/A'],'0')
testset['University Degree'] = testset['University Degree'].replace(['0','#N/A'], 'No')
testset['Hair Color'] = testset['Hair Color'].replace(['0','#N/A'], 'Unknown')
testset.head()

#ONEHOTENCODE

#DATASET
dataset=pd.get_dummies(dataset,columns=['Gender'],prefix=['Gender'])
dataset=pd.get_dummies(dataset,columns=['University Degree'],prefix=['University Degree'])
dataset=pd.get_dummies(dataset,columns=['Hair Color'],prefix=['Hair Color']) 
dataset=pd.get_dummies(dataset,columns=['Year of Record'],prefix=['Year of Record']) 
#dataset=pd.get_dummies(dataset,columns=['Country'],prefix=['Country']) 

#TESTSET
testset=pd.get_dummies(testset,columns=['Gender'],prefix=['Gender'])
testset=pd.get_dummies(testset,columns=['University Degree'],prefix=['University Degree'])
testset=pd.get_dummies(testset,columns=['Hair Color'],prefix=['Hair Color']) 
testset=pd.get_dummies(testset,columns=['Year of Record'],prefix=['Year of Record']) 
#testset=pd.get_dummies(testset,columns=['Country'],prefix=['Country']) 
dataset.head()
testset.head()

#REMOVING OUTLIERS

#Splitting data 
X=dataset.drop(['Income in EUR'], axis=1)
Y=dataset['Income in EUR']

xtrain,xvalidate,ytrain,yvalidate=train_test_split(X,Y,test_size=0.2,random_state=0)
x_test=testset.drop("Income",axis=1)

#Applying model
regressor=BayesianRidge()
regressor.fit(xtrain,ytrain)
y_predict=regressor.predict(xvalidate)
result=regressor.predict(x_test)
res=pd.DataFrame(x_test['Instance'])
res['Income']=result
res.index=x_test.index
res.to_csv("predictions_final.csv")
rms=np.sqrt(mean_squared_error(yvalidate,y_predict))
print("rmes is "+ str(rms))
```
